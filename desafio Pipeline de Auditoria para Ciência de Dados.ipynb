```python
filial = str(input("Digite o nome da sua filial: "))

valor = 1

valores_com_ruido = []

lista_limpa = []

def limpar_vendas(parametros):
    for i in valores_com_ruido:
        if i > 0 and i < 10000:
            lista_limpa.append(i)
        
def analisar_dados(parametros):
    faturamento_total = sum(lista_limpa)  
    media_de_vendas = faturamento_total / len(lista_limpa)
    return faturamento_total, media_de_vendas
    
def definir_status (media):
    media_tupla = media[1]
    if media_tupla < 2000:
        performance = "Performance Crítica"
        
    elif media_tupla >= 2000 and media_tupla < 5000:
        performance = "Performance Estável"
        
    else:
        performance = "Alta Performance"
        
    return performance
    
while valor != 0 :
    valor = float(input("digite o valor da venda: "))
    if valor < 0:
        while valor <= 0:
            valor = float(input("digite o valor da venda novamente: "))
    if valor == 0:
        break
    
    if valor > 0:
        valores_com_ruido.append (valor)
        
limpeza = limpar_vendas(valores_com_ruido)

total_media = analisar_dados(limpeza)

status = definir_status(total_media)



print(lista_limpa)

print('-'*40)

print(f"total e media: {total_media}")

print('-'*40)

print(status)


```
